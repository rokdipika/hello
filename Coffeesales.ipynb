## **Data Visualisation through seaborn**

```python
sns.lineplot(x="Time_of_Day",y="coffee_name",data= df)
# title and labeling 
 plt.xlabel("Time_of_Day")
 plt.ylabel("coffee_name")
 plt.title("coffee_name by Time_of_Day")
 # show result
 plt.show()
```

```python
 sns.hist(df["hour_of_day"], bins= 10, kde= True)
 plt.xlabel("hour_of_day")
 plt.ylabel("Frequency")
 plt.show()
```

```python
# Histogram
plt.hist(df["Time_of_Day"] ,color="skyblue", edgecolor="black")
plt.title("coffee_name Distribution")
plt.xlabel("Time_of_Day")
plt.ylabel("coffee_name")
plt.show()
```

```python
sns.barplot(x="hour_of_day",y="coffee_name",data=df)
# title and labeling
 plt.title("bar plot between name of coffee and time of the day")
 plt.xlabel("time of the day")
 plt.ylabel("name of the coffee")
 # show result
 plt.show()
```

```python
sns.countplot(x="Weekday",hue= "Weekday",data= df)
# title and labeling
 plt.title("Count plot of Weekdays")
 plt.xlabel("days of the week")
 
 # show result
 plt.show()
```
